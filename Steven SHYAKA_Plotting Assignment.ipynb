```python
# YOUR CODE HERE

print(df['co2'].head(10))
co2_emission=df['co2']

'''
The code below uses seaborn and plots the data from the co2 column into a histogram
Bin size of 20 provides a good view

'''
sns.set(style="whitegrid")
sns.histplot(data=co2_emission,color='blue',edgecolor='black',bins=20)
plt.xlabel('CO2 Emissions')
plt.ylabel('Frequency')
plt.title('Distribution of Carbon Dioxide Emissions')
plt.savefig('co2_Emission.png')
plt.show()
```

**INTERPRETATION**  
The x-axis represents CO2 emissions, measured in some unit. The values range from 0 to 700.The y-axis indicates the frequency of occurrences for each emission level, ranging from 0 to 600.The histogram displays a unimodal distribution with a single prominent peak. This suggests that most of the data points fall around a central range of CO2 emissions. The peak of the histogram occurs in the range of approximately 200 to 300 CO2 emissions. This is where the highest frequency of CO2 emission values is observed, indicating that this is the most common range of emissions in the dataset.The distribution appears to be slightly right-skewed, as the tail on the right side of the histogram (higher emission values) extends further than the tail on the left. This skewness indicates that there are some higher emission values that are less frequent but still present.The histogram shows that lower emissions (0-100) and very high emissions (500-700) are less frequent. The bulk of the emissions are concentrated in the middle ranges, particularly around the peak.

```python
'''
The code below calculates the correlation matrix of the dataset and then creates a heatmap using seaborn's heatmap function.
The heatmap visualizes the correlations between different columns in the given dataset,
where values closer to 1 or -1 indicate stronger positive or negative correlations, respectively.
'''
numeric_columns = df.select_dtypes(include=['number'])
corr_matrix = numeric_columns.corr()

plt.figure(figsize=(9, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap of the Fuel Economy Data')
plt.savefig('heatmap.png')
plt.show()
```
